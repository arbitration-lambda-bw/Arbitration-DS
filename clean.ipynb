```python
def get_n_rows(file, n_rows):
    with open(file, 'rt') as csvf:
        r = csv.reader(csvf)
        headers = next(r)
        data = []
        row_count = sum(1 for row in r)
        rrows = [random.randint(1,row_count) for i in range(n_rows)].sort()
    with open(file, 'rt') as csvf:
        r = csv.reader(csvf)
        next(r)
        for i in range(n_rows):
            data.append(next(r))
    return(headers, data)
```

```python
def nrows_to_df(file, n_rows):
    rt = sum(1 for lin in open(file, 'rt'))
    skip = sorted(random.sample(range(1,rt+1), rt - n_rows))
    df = pd.read_csv(file, skiprows = skip)
    return(df)
```
